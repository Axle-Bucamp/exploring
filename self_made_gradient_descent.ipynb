```python
# cost = 1/n sum( ( y - tanh( sum(x*w + b)) )^2 )
# neuron = tanh(sum(x*w + b))

# d cost / dw =  d/w 1/2n sum( ( tanh( sum(x*w + b)) )^2 - y )
## sum rule
## d cost / dw = 1/2n sum( d/dw (u - v)^2 )
### chaine rule
### d cost / dw = 1/2n 2(u - v) *  d/dw (u - v)
### d cost / dw = 1/2n 2(u - v) *  (d/dw u - d/dw v)
### d cost / dw = -1/n (u - v) * (d/dw u)
# and
### d cost / db = -1/n (u - v) * (d/db u)

# d/dw u =  1 - tanh(x)^2
# d/db u =  1 - tanh(1)^2

### gradient = -1/n (u - v) * (d/db u) + -1/n (u - v) * (d/dw u) + 2y


import numpy as np
import plotly.graph_objects as go
import math
x = np.array(x) / 20
grad = []
for i in range(20):
  arr = []
  for z in range(20):
    # x[i] # y[z]
    for w in range(20):
      for b in range(20):
        arr.append(   2*y[z] * (1 - math.tanh(w/20*x[i] + b/20 ) ) * x[i] - ( y[z] - math.tanh( w/20*x[i] + b/20 ) ) *  (1- math.tanh(w/20*x[i] + b/20 )**2 )  )
  grad.append(arr)

fig = go.Figure(data=[go.Surface(z=grad)])

fig.update_layout(title='Gradient Function of sum(y - than(wx+b)^2 )/n', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
fig.write_html("htmlintegration.html")
```

```python
# voir corriger l'explosion de gradient ?
import numpy as np
class layer():
  w = []
  b = []
  linearized = []
  activated = []

  graderr = []
  inputvect = []
  derivative = []

  activation = lambda x : x

  def __init__(self, Uactivation,size):
    self.activation = Uactivation
    mu, sigma = 0, 0.1
    self.w = np.random.normal(mu, sigma, size)
    self.b = np.random.normal(mu, sigma, size[0])

  def learn(self,grad, learning_rate,w,):
    self.w = self.w - learning_rate*grad[0]
    self.b = self.b - learning_rate*grad[1]

  def derivate_tanhMethod(self,input):
    self.derivative = np.array(( 1 - np.tanh(self.w.T * input + self.b)**2) * input )
    return self.derivative

  def activate(self,x):
    self.activated = np.array(self.activation(x))
    return self.activated

  def lineare(self,x):
    self.inputvect = x
    self.linearized = np.array(np.sum(self.w.T * x,axis=1) + self.b)
    return self.lineare

  def predict(self,x):
    self.lineare(x)
    return self.activate(self.linearized)
```

```python
def train(network, x,y,learning_rate) :
  y_pred = x.copy()
  for layer in network:
    y_pred = layer.predict(y_pred)

  ## learning phase
  print("error : "+ str(np.sum((y -y_pred)**2 )))
  for layer in reversed(range(len(network))):

    if layer == 0:
      network[layer].derivate_tanhMethod(x)
    else :
      network[layer].derivate_tanhMethod(network[layer].activated)

    if layer == len(network) -1:
      network[layer].graderr = network[layer].derivative * (y - y_pred)
    else :
      network[layer].graderr = network[layer].derivative * np.sum(network[layer+1].w * network[layer+1].graderr)

  for layer in range(len(network)):
    if layer == 0:
      network[layer].w = network[layer].w - learning_rate * network[layer].graderr * x
      network[layer].b = network[layer].b - learning_rate * network[layer].graderr * x
    else :
      network[layer].w = network[layer-1].w - learning_rate * network[layer].graderr *network[layer-1].activated
      network[layer].b = network[layer-1].b - learning_rate * network[layer].graderr *network[layer-1].activated
```

```python
tanh = lambda x : np.tanh(x)
network = []
# size = (neurone, entre ), sortie = neurone
network.append(layer(Uactivation=tanh, size=(20,20) ))
network.append(layer(Uactivation=tanh, size=(20,20) ))
network.append(layer(Uactivation=tanh, size=(1,20) ))

train_step = 10
for i in range(train_step):
  train(network, np.array(x),np.array(y)[0], 0.1)
```
